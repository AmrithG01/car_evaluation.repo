```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('car.data')
df.head(5)
```

```python
col_name = ['buying','maint','doors','person','lug_boost','safety','class']
df.columns = col_name
df
```

```python

df.describe()
```

```python
for c in col_name:
    print(df[c].value_counts())
    print('-----'*15)
```

```python
df[df.duplicated()]
#to get df with all duplicate values
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()

for c in col_name:
    df[c] = oe.fit_transform(df[[c]])

df
```

```python
#to make input and out format use iloc

X = df.iloc[: , 0:-1]
y = df.iloc[ : , -1]
# here row , coumn splitting
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=1)
rf.fit(X_train, y_train)      
y_pred = rf.predict(X_test) 
```

```python
from sklearn.metrics import accuracy_score , precision_score,roc_auc_score,roc_curve

accuracy = accuracy_score(y_test,y_pred)
print(f'accuracy for random forest model:{accuracy}')
```

```python
# to know fearure importnace
imp = rf.feature_importances_

feature_imp = pd.DataFrame({'Feature': X_train.columns, 'Importance': imp})

print(feature_imp.sort_values(by='Importance',ascending=False))


```

```python
import pickle
pickle.dump(rf, open("car_model.pkl", "wb"))
# for file creation in binary write mode
```

```python
import streamlit as st
model = pickle.load(open("car_model.pkl","rb"))
st.title("Car Evaluation Prediction")

buying = st.selectbox("Buying Price", ["low","med","high","vhigh"])
maint = st.selectbox("Maintenance", ["low","med","high","vhigh"])
doors = st.selectbox("Doors", [2,3,4,5])
persons = st.selectbox("Persons Capacity", [2,4,"more"])
lug_boot = st.selectbox("Luggage Boot", ["small","med","big"])
safety = st.selectbox("Safety", ["low","med","high"])

if st.button("Predict"):
    input_data = np.array([[buying,maint,doors,persons,lug_boot,safety]])
    prediction = model.predict(input_data)
    st.success(f"Car Acceptability: {prediction[0]}")
```
